```python
import pandas as pd
import re

url_regex = r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?¬´¬ª‚Äú‚Äù‚Äò‚Äô]))"       
```

```python
data = pd.read_csv('dataset/ask_play_station.csv')
data['User'][0]
```

```python
def RemoveURLsFromData():
    pom = 0
    for index, row in data.iterrows():
        pom = pom+1
        row[0] = re.sub(url_regex, "URL_POSITION", str(row[0]))
        row[1] = re.sub(url_regex, "URL_POSITION", str(row[1]))
        #if(pom == 100):
            #break
```

```python
def ValidateURLsFilter():
    for column in data:
        pom = 0
        for instance in data[column]:
            pom = pom+1
            #if(pom == 100):
                #break
            testUrl = re.findall(url_regex, str(instance))
            if len(testUrl) != 0:
                raise Exception("URL filter didn't catch an URL")
        #break
```

```python
RemoveURLsFromData()
ValidateURLsFilter()
```
