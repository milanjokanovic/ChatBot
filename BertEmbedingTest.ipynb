```python
from simpletransformers.language_representation import RepresentationModel
from simpletransformers.config.model_args import ModelArgs
```

```python
model_args = ModelArgs(max_seq_length=60)
model = RepresentationModel(
        model_type="bert",
        model_name="bert-base-uncased",
        args=model_args,
        use_cuda=False
    )
```

```python
sentences = ["Machine Learning and Deep Learning are part of AI", "Machine Science will excel in future"]
word_vectors = model.encode_sentences(sentences, combine_strategy=None)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("dataset/ask_play_station.csv")
df.dropna(inplace = True)
df
```

```python
avarage_size = df['Company'].apply(len).mean()
avarage_size
```

```python
max = 0
for el in df['Company']:
    if(len(el) > max):
        max = len(el)
print(max)
```

```python
df["User"] = df["User"].str.pad(max, side ='left')
```

```python
len(df['User'][0])
```

```python
train_data, test_data = train_test_split(df, test_size=0.3, random_state=42)
```
