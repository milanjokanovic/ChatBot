```python
companyData = data.loc[(data['author_id'].astype(str).str[0].str.match("\w(?<![\d_])"))]
selected_companies = companyData['author_id'].value_counts()[0:20]
company_list = list(selected_companies.keys())
data_for_use = data[(data['author_id'].isin(company_list) | data['text'].isin(company_list))]
data_for_use
```

```python
from langdetect import detect
list = []

for el in data_for_use['text'].str[:]:
    try:
        if(detect(el) == 'en'):
            list.append(el)
    except:
        pass
```

```python
data_for_use = data[(data['author_id'] == 'AmazonHelp') & (data['text'].isin(list))].dropna(subset=['in_response_to_tweet_id'])
```

```python
data_for_use = data[(data['author_id'] == 'AskPlayStation') & (data['text'].isin(list))].dropna(subset=['in_response_to_tweet_id'])
```

```python
data_for_use = data[(data['author_id'].isin(company_list)) & (data['text'].isin(list))].dropna(subset=['in_response_to_tweet_id'])
```

```python
conversation_data = []
for index, row in data_for_use.iterrows():
    filter_value = row['in_response_to_tweet_id']
    value = data[data['tweet_id'] == filter_value]
    element = ""
    for ind, el in value.iterrows():
        element = el['text']
    if(isinstance(element, str)):
        conversation_data.append([element, row['text']])
header = ["User", "Company"]
wrangled_data = pd.DataFrame(conversation_data, columns=header)
```

```python
wrangled_data.to_csv('dataset/wrangled_data.csv', index=False)
```
